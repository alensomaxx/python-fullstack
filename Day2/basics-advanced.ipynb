### MATCH
<p>Kinda similar to Switch but no need of any break statement </p>

```python
a = input()
match a:
    case "hello":
        print("Hi!")
    case "bye":
        print("Goodbye!")
    case _:
        print("I don't understand.")
```

```python
a = input()
if a in "aeiouAEIOU":
    print("Vowel")
else:
    print("Consonant")
```

```python
for i in "abcdefghi":
    print (i)
    continue # this will skip to the next iteration
print("hello world")
```

```python
x = lambda a,b:a+b
print(x(5,10), x(2,3))
```

```python
l = [[2,3],[1,4,7],[5,0],[9,8,6]]
l.sort(key=lambda x: x[-1]*x[-2],reverse=True)
print(l)
```
