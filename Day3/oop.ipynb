### String Manipulation

```python
s = "Python programming is fun"
words = s.split()
print (words)
reversed_words = words[::-1]
print (reversed_words)
result = ' '.join(reversed_words)
print (result)
```

```python
s = "Python programming is fun"
print(s)
print(s.upper())
print(s.lower())
print(s.replace("fun", "awesome"))
```
